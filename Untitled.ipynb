```python
# Cell 1: Imports & settings
import os
import glob
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, GRU, Dense, Dropout, BatchNormalization, Concatenate, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import joblib
np.random.seed(42)
tf.random.set_seed(42)

# User-editable
DATA_DIR = "./data"   # <- put train_FD001.txt etc. here
SEQ_LEN = 50
CAP_RUL = 125
BATCH_SIZE = 128
EPOCHS = 80
MODEL_SAVE_PATH = "grudomain_fd_all.h5"
SCALER_PATH = "scaler_fd_all.joblib"
SENSOR_COLS_PATH = "sensor_cols_fd_all.joblib"
```

```python
# Cell 2: helper to read CMAPSS style whitespace-delimited file and show files
def read_cmapss(filepath):
    col_names = (["unit", "cycle",
                  "op_setting_1", "op_setting_2", "op_setting_3"] +
                 [f"s{i}" for i in range(1, 22)])  # s1..s21
    df = pd.read_csv(filepath, sep='\s+', header=None, names=col_names)
    return df

# list files
train_files = sorted(glob.glob(os.path.join(DATA_DIR, "train_FD*.txt")))
test_files  = sorted(glob.glob(os.path.join(DATA_DIR, "test_FD*.txt")))
rul_files   = sorted(glob.glob(os.path.join(DATA_DIR, "RUL_FD*.txt")))

print("train files:", train_files)
print("test files:", test_files)
print("rul files:", rul_files)
```

```python
# Cell 3: load train files into dict and add RUL column
train_dfs = {}
for tfp in train_files:
    key = os.path.basename(tfp).replace("train_", "").replace(".txt", "")  # e.g., FD001
    df = read_cmapss(tfp)
    # add RUL = max_cycle_for_unit - cycle
    max_cycle = df.groupby("unit")["cycle"].transform("max")
    df["RUL"] = max_cycle - df["cycle"]
    train_dfs[key] = df
    print(f"Loaded {key}: shape={df.shape}, units={df['unit'].nunique()}")
```

```python
# Cell 4: detect sensor columns and fit a global scaler on combined train sensor+op columns
example_key = next(iter(train_dfs))
example_df = train_dfs[example_key]
sensor_cols = [c for c in example_df.columns if c.startswith("s") or c.startswith("op_setting")]
print("Sensor+op columns detected:", sensor_cols)

# combine all train frames to fit scaler
combined_train = pd.concat([df[sensor_cols] for df in train_dfs.values()], axis=0)
scaler = StandardScaler().fit(combined_train.values)
print("Fitted global scaler on combined train data:", combined_train.shape)
```

```python
import joblib

# Save scaler
joblib.dump(scaler, SCALER_PATH)
print(f"Scaler saved to: {SCALER_PATH}")

# Save sensor column list
joblib.dump(sensor_cols, SENSOR_COLS_PATH)
print(f"Sensor columns saved to: {SENSOR_COLS_PATH}")
```

```python
# Cell 5: Preprocessing function that works for train and test
def preprocess_df_for_model(df, sensor_cols, scaler, cap_rul=CAP_RUL, is_train=True):
    df = df.copy()
    # If training, ensure RUL exists; if not present (test) skip RUL creation
    if is_train:
        if "RUL" not in df.columns:
            # If train file lacks RUL (unexpected), compute via max-cycle method
            max_cycle = df.groupby("unit")["cycle"].transform("max")
            df["RUL"] = max_cycle - df["cycle"]
    # cap RUL if present
    if "RUL" in df.columns:
        df["RUL_capped"] = df["RUL"].clip(upper=cap_rul)
    # apply scaler to sensor+op columns
    df[sensor_cols] = scaler.transform(df[sensor_cols].values)
    return df
```

```python
# Cell 6: generate sequences, appending domain one-hot vector as extra features at every timestep
def generate_sequences_with_domain(df, sensor_cols, seq_len=SEQ_LEN, domain_id=0, num_domains=4, label_col="RUL_capped"):
    Xs = []
    ys = []
    units = sorted(df["unit"].unique())
    # one-hot domain vector
    domain_vec = np.zeros(num_domains, dtype=np.float32)
    domain_vec[domain_id] = 1.0
    for u in units:
        tmp = df[df["unit"] == u].sort_values("cycle")
        data = tmp[sensor_cols].values  # (n_cycles, n_features)
        n_cycles = data.shape[0]
        if n_cycles < seq_len:
            # pad by repeating first row
            pad_n = seq_len - n_cycles
            pad = np.vstack([data[0]] * pad_n)
            data_padded = np.vstack([pad, data])
            X = data_padded
            if label_col in tmp.columns:
                y = tmp[label_col].values[-1]  # last available label
                Xs.append(np.hstack([X, np.tile(domain_vec, (seq_len,1))]))
                ys.append(y)
            continue
        for start in range(0, n_cycles - seq_len + 1):
            end = start + seq_len
            X = data[start:end]
            # append domain vec to each timestep
            X_aug = np.hstack([X, np.tile(domain_vec, (seq_len,1))])
            Xs.append(X_aug)
            if label_col in tmp.columns:
                ys.append(tmp[label_col].values[end-1])
    X = np.array(Xs, dtype=np.float32)
    y = np.array(ys, dtype=np.float32) if len(ys) > 0 else None
    return X, y
```

```python
# Cell 7: create sequences for each FD train, add domain id mapping, and combine
domain_keys = sorted(train_dfs.keys())  # e.g. ['FD001','FD002',...]
domain_map = {k:i for i,k in enumerate(domain_keys)}
print("Domains:", domain_map)

X_list = []
y_list = []
domain_counts = {}
for key, df in train_dfs.items():
    did = domain_map[key]
    df_proc = preprocess_df_for_model(df, sensor_cols, scaler, cap_rul=CAP_RUL, is_train=True)
    Xk, yk = generate_sequences_with_domain(df_proc, sensor_cols, seq_len=SEQ_LEN, domain_id=did, num_domains=len(domain_keys), label_col="RUL_capped")
    print(f"{key}: sequences={Xk.shape[0]}, each shape={Xk.shape[1:]} (features include domain one-hot)")
    X_list.append(Xk)
    y_list.append(yk)
    domain_counts[key] = Xk.shape[0]

# combine
X_all = np.vstack(X_list)
y_all = np.hstack(y_list)
print("Combined X shape:", X_all.shape, "y shape:", y_all.shape)
```

```python
# Cell 8: train/val split ensuring domains are proportionally represented
# We'll create an array of domain ids per sequence so splitting keeps proportions using stratify
# Recreate domain ids per sequence
domain_ids = []
for key, Xk in zip(domain_keys, X_list):
    domain_ids.extend([domain_map[key]] * Xk.shape[0])
domain_ids = np.array(domain_ids)

X_train, X_val, y_train, y_val, d_train, d_val = train_test_split(
    X_all, y_all, domain_ids, test_size=0.15, random_state=42, stratify=domain_ids)

print("Train shape:", X_train.shape, y_train.shape, "Val shape:", X_val.shape, y_val.shape)
```

```python
# Cell 9: build model
def build_strong_gru(input_shape, gru_units=[256,128], dropout_rate=0.3):
    inp = Input(shape=input_shape, name="seq_input")  # (SEQ_LEN, n_features+domain)
    x = inp
    # stacked GRUs
    for i, units in enumerate(gru_units):
        return_seq = (i < len(gru_units)-1)
        x = GRU(units, return_sequences=return_seq, name=f"gru_{i+1}")(x)
        x = BatchNormalization()(x)
        x = Dropout(dropout_rate)(x)
    # final dense head
    x = Dense(128, activation="relu")(x)
    x = Dropout(dropout_rate)(x)
    out = Dense(1, activation="linear", name="rul_output")(x)
    model = Model(inputs=inp, outputs=out)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4), loss="mse", metrics=["mae"])
    return model

input_shape = (SEQ_LEN, X_train.shape[2])
model = build_strong_gru(input_shape, gru_units=[256, 128], dropout_rate=0.35)
model.summary()
```

```python
# Cell 10: training with callbacks
callbacks = [
    EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True, verbose=1),
    ModelCheckpoint(MODEL_SAVE_PATH, monitor="val_loss", save_best_only=True, verbose=1)
]

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    verbose=2
)
```

```python
# Cell 11: plot training/validation loss
plt.figure(figsize=(8,4))
plt.plot(history.history["loss"], label="train_loss")
plt.plot(history.history["val_loss"], label="val_loss")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.legend()
plt.title("Training History (combined FD datasets)")
plt.show()
```

```python
# Fixed Cell 12: Evaluate per FD test set (robust loading and clipping of RUL file)
def build_test_X_for_model(test_df, sensor_cols, scaler, seq_len, domain_id, num_domains):
    df_proc = preprocess_df_for_model(test_df, sensor_cols, scaler, is_train=False, cap_rul=CAP_RUL)
    # For each unit, take last seq_len cycles (pad front if shorter) and append domain one-hot
    X_test = []
    units = sorted(df_proc["unit"].unique())
    domain_vec = np.zeros(num_domains, dtype=np.float32); domain_vec[domain_id] = 1.0
    for u in units:
        tmp = df_proc[df_proc["unit"]==u].sort_values("cycle")
        arr = tmp[sensor_cols].values
        if arr.shape[0] >= seq_len:
            arr_used = arr[-seq_len:]
        else:
            pad_n = seq_len - arr.shape[0]
            pad = np.vstack([arr[0]] * pad_n)
            arr_used = np.vstack([pad, arr])
        X_test.append(np.hstack([arr_used, np.tile(domain_vec, (seq_len,1))]))
    return np.array(X_test, dtype=np.float32), units

results = {}
for test_fp in test_files:
    key = os.path.basename(test_fp).replace("test_", "").replace(".txt", "")  # e.g., FD001
    print("\nProcessing test set:", key)
    test_df = read_cmapss(test_fp)
    # find corresponding rul file
    rul_fp = os.path.join(DATA_DIR, f"RUL_{key}.txt")
    if not os.path.exists(rul_fp):
        print("  missing RUL file for", key)
        continue
    domain_id = domain_map.get(key, None)
    if domain_id is None:
        print("  no domain mapping for", key)
        continue

    # build X_test and list of unit ids (sorted)
    X_test, test_units = build_test_X_for_model(test_df, sensor_cols, scaler, SEQ_LEN, domain_id, num_domains=len(domain_keys))

    # load true RULs robustly and cap them using numpy.clip
    y_test_true = pd.read_csv(rul_fp, sep='\s+', header=None).values.ravel()
    # Cap using np.clip (works across numpy versions)
    y_test_true = np.clip(y_test_true, a_min=None, a_max=CAP_RUL)

    # Sanity check: lengths should match
    if len(y_test_true) != len(test_units):
        print(f"  Warning: number of RUL entries ({len(y_test_true)}) != number of test units ({len(test_units)}).")
        # Try to proceed but slice/extend as needed (prefer slicing to avoid misalignment)
        min_len = min(len(y_test_true), len(test_units))
        y_test_true = y_test_true[:min_len]
        X_test = X_test[:min_len]
        test_units = test_units[:min_len]

    # Predict and evaluate
    y_test_pred = model.predict(X_test).ravel()
    rmse = math.sqrt(mean_squared_error(y_test_true, y_test_pred))
    print(f"  Test units: {len(test_units)}, RMSE (capped to {CAP_RUL}): {rmse:.4f}")
    results[key] = {"rmse": rmse, "y_true": y_test_true, "y_pred": y_test_pred, "units": test_units}
```

```python
# Cell 13: plots for each FD test set
for key, res in results.items():
    y_true = res["y_true"]
    y_pred = res["y_pred"]
    plt.figure(figsize=(6,4))
    plt.scatter(y_true, y_pred, alpha=0.6)
    plt.plot([0, CAP_RUL],[0,CAP_RUL],"k--", linewidth=1)
    plt.xlim(0, CAP_RUL); plt.ylim(0, CAP_RUL)
    plt.xlabel("True RUL"); plt.ylabel("Predicted RUL")
    plt.title(f"{key}: True vs Predicted RUL (RMSE={res['rmse']:.3f})")
    plt.show()
```

```python
# Cell 14: thresholding & sample alerts
WARNING_THRESHOLD = 30
CRITICAL_THRESHOLD = 10

def get_alert(rul_value):
    if rul_value <= CRITICAL_THRESHOLD:
        return "CRITICAL"
    elif rul_value <= WARNING_THRESHOLD:
        return "WARNING"
    else:
        return "NORMAL"

for key, res in results.items():
    print(f"\nAlerts for {key} (first 10 units):")
    for i in range(min(10, len(res["units"]))):
        unit = res["units"][i]
        true_r = res["y_true"][i]
        pred_r = res["y_pred"][i]
        print(f" Unit {unit}: True={true_r:.1f}, Pred={pred_r:.1f}, Alert={get_alert(pred_r)}")
```
